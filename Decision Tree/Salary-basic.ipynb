```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
df.head()
```

```python
sns.countplot(x='salary_more_then_100k',hue='company',data=df)
```

```python
inputs=df.drop('salary_more_then_100k',axis=1)
target=df['salary_more_then_100k']
```

```python
target
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
comp=LabelEncoder()
jobb=LabelEncoder()
deg=LabelEncoder()
```

```python
inputs['new_comp']=comp.fit_transform(inputs['company'])
inputs['new_job']=jobb.fit_transform(inputs['job'])
inputs['new_degree']=deg.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs.drop(['company','job','degree'],axis=1,inplace=True)
```

```python
inputs
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```

```python
model.predict([[1,1,0]])
```
